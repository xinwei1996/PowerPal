```python
import io
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def extract_state(df):
    df['coords'] = df.apply(lambda row: f"{row['latitude']}, {row['longitude']}", axis=1)
    df['location'] = df.apply(lambda row: geolocator.reverse(row.coords, exactly_one=True, timeout=None), axis=1)
    return df
```

```python
power_plant_aus = extract_state(power_plant_aus)
```

```python
state_value = list()
for index, row in power_plant_aus.iterrows():
    if 'state' in row.location.raw['address']:
        val = row.location.raw['address'].get('state', '')
    else:
        val = row.location.raw['address'].get('territory', '')
        
    state_value.append(val)
```

```python
power_plant_aus['state'] = state_value
```

```python
# Displaying the unique state values
power_plant_aus.state.unique()
```

```python
# Drop the location and coords attribute
updated_power_plant_aus = power_plant_aus.drop(['location', 'coords'], axis = 1)
```
