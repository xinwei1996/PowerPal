#  Cleaning Data for user story 1.3 & 1.2

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load clean data
all_aps = pd.read_csv("all_appliance.csv")
```

```python
all_aps
```

```python
# expand rows of the manufacturing country
all_aps['Manufacturing_Country'] = all_aps['Manufacturing_Country'].str.split(',')
df = all_aps.explode('Manufacturing_Country')




```

```python
# filter rows with energy rating of - (not useful to us)
df = df[~df["Energy_Rating"].str.contains("-")]
```

```python

# change column to numeric
df["Energy_Rating"] = pd.to_numeric(df["Energy_Rating"], downcast="float")
# find the average energy rating of each applaince.
avg_ER = df.groupby(["Appliance_Category"])["Energy_Rating"].mean()
```

```python
# add avg Energy rating into a df (DONE FOR US1.2)
av_ER_df = avg_ER.to_frame(name = "avg_Energy_Rating" ).reset_index()
```

```python
# find the number of countries 
countries_df =  df.groupby(['Manufacturing_Country', "Appliance_Category"]).agg(
    Mean=('Energy_Rating', np.mean),
    Count=('Appliance_Category', np.size)).reset_index()
```
